```python
from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage, SystemMessage

# Initialize the model
llm = ChatOllama(model="llama3.1:8b", temperature=0.7)
print("Model initialized!")
```

```python
%%time 

print("Test 1")
response = llm.invoke("What are the steps to start career planning?")
print(response.content)
```

```python
%%time

prompt = """
I am a Computer Science student with 3 years of experience in Python, 
but I have zero experience in Cloud Computing.
Suggest 3 specific steps I should take to become a Backend Engineer.
Keep it brief.
"""

print("Test 2")
response = llm.invoke(prompt)
print(response.content)
```

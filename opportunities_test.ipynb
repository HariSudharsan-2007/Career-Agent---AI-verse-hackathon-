```python
from duckduckgo_search import DDGS
import ollama

MODEL = "llama3.1:8b"

def find_opportunities(topic, category="internship"):
    # category can be: internship, certification, or course
    print(f"Looking for {category}s in '{topic}'...")

    # Construct a targeted query
    if category == "internship":
        query = f"latest {topic} internships for students India apply online 2024 2025"
    elif category == "certification":
        query = f"best recognized {topic} certifications for beginners 2025 with certificate"
    else:
        query = f"free {topic} courses with projects"

    # Search
    try:
        results = DDGS().text(query, max_results=6)
    except Exception as e:
        print(f"Search error: {e}")
        return

    # Filter and Format for Agent
    if not results:
        print("No opportunities found.")
        return

    context = "\n".join([f"Title: {r['title']}\nSnippet: {r['body']}\nLink: {r['href']}" for r in results])
    
    # LLM Summary
    print(f"Found {len(results)} listings. Summarizing...")
    
    prompt = f"""
    You are a Career Mentor.
    The user is looking for {category}s in {topic}.
    Summarize the top 3 opportunities from the search results below.
    
    For each opportunity, list:
    1. Title
    2. Key Requirement or Benefit
    3. The Link
    
    SEARCH RESULTS:
    {context}
    """

    response = ollama.chat(model=MODEL, messages=[{'role': 'user', 'content': prompt}])
    print("OPPORTUNITIES FOUND:")
    print(response['message']['content'])

if __name__ == "__main__":
    # Test 1: Find Python Internships
    find_opportunities("Python", "internship")
    
    # Test 2: Find AWS Certifications
    find_opportunities("AWS Cloud", "certification")
```
